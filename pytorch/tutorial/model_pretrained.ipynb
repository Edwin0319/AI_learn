```python
import torchvision
from torchvision.models import VGG16_Weights, vgg16
import os
import torch.nn as nn
import torch
```

vgg16_false是没有訓練過的模型,没有優化參數
vgg16_true是優化過的參數的模型

```python

os.environ["TORCH_HOME"] = "./models" # 修改下载位置
# train_data = torchvision.datasets.ImageNet("./data_image_net", split="train", download=True, transform=torchvision.transforms.ToTensor())
vgg16_false = vgg16(weights=None, progress=True)
vgg16_true = vgg16(weights=VGG16_Weights.DEFAULT, progress=True)

# torch.save(vgg16_true, './models/vgg16_true.pth')
# torch.save(vgg16_false, './models/vgg16_false.pth')
print(vgg16_true)
print(vgg16_false)
```

```python
print(vgg16_true)
print(vgg16_false)
print("altered")
train_data = torchvision.datasets.CIFAR10('./dataset', train=True, transform=torchvision.transforms.ToTensor(), download=True)
vgg16_true.classifier.add_module('add_linear', nn.Linear(1000, 10)) # 可以直接在模型加操作
print(vgg16_true)
vgg16_false.classifier[6] = nn.Linear(1000, 10) # 也可以直接改模型
print(vgg16_false)
```
